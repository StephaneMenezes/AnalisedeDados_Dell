```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro05/oficina/arquivos/dadosoficina05.txt')
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribuição dos valor de alugueis por imóveis')
sns.kdeplot(df['rent amount (R$)'],color = 'r',shade = True)
plt.show()
```

O valores mais comuns são : Aproximadamente 2500

```python
plt.figure(figsize=(10,5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribuição dos valor de alugueis por imóveis')
sns.kdeplot(df['fire insurance (R$)'],color = 'orange', shade = True)
plt.show()
```

Os valores mais incomuns são acima de 150
