```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
filmes_uri = "https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json"  #Note que arquivo está em formato Json por isso armazenamos o endereço em uma variavel
filmes = pd.read_json(filmes_uri) #E depois criamos o dataframe
filmes.head()
```

```python
filmes.min()
```

## 1 - Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

```python
plt.figure(figsize = (6,10))
plt.title("Orçamento por Filme")
sns.barplot(y="Film" , x='Budget_M', color = "r", ci=None, data=filmes)
plt.show()
```

* Segundo o gráfico o filme com maior orçamento foi: Pirates of the Caribbean: At World's End

## 2 - Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

```python
plt.figure(figsize = (6,10))
plt.title("Bilheteria por Filme")
sns.barplot(y="Film" , x='Worldwide_Gross_M', ci=None, data=filmes)
plt.show()
```

 # Segundo o gráfico o filme com : 
   * Maior bilheteria foi: Avatar 
   * Menor bilheteria foi: Iron Man 2 

# 3 - Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

```python
plt.figure(figsize=(13,5))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue= 'Genre', data=filmes)
plt.show()
```

## Segundo o gráfico o ano em que:
   * Foram feitos menos filmes de ação foi: 2010
   * Foram feitos menos filmes de animação foi: 2008 e 2011. Ambos com as mesmas quantidades (1 filme) 

# Crie uma visualização que mostre a média de orçamento de filmes por gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento?

```python
plt.figure(figsize=(15,8))
plt.title('Média de orçamento por gênero em cada ano')
sns.barplot(x="Year", y ='Budget_M' , ci = None, hue= 'Genre', data=filmes)
plt.show()
```

## Segundo o gráfico o ano em que:
   * Os filmes de ação tiveram maior orçamento: 2007
   * Os filmes de animação tiveram maior orçamento: 2008
