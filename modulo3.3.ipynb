```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt')
df.head()
```

```python
plt.figure(figsize=(20,5))
plt.title('Variação da temperatura ao longo do tempo')
sns.lineplot(x="Month", y="Sales", data=df)
plt.show()
```
