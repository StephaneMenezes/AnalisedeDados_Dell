```python
import pandas as pd                 
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
plt.figure(figsize=(8,5))
sns.set(style="darkgrid")
plt.title('Relação entre a aceleração e o consumo de combustível em diferentes países')
sns.scatterplot(x="acceleration", y="mpg", data=mpg, hue = 'origin')
```

* Sabendo disso, existe uma correlação entre a aceleração e o consumo? 
R = Sim, podemos observar que existe uma relação positiva, pois quando a velocidade cresce o consumo de combustivel tbm. 
* De que tipo? 
R = Positiva 
* A correlação é a mesma para os diferentes países de origem?
R = Sim, esse aumento ocorre em todos os países mas a USA se destaca pela quantidade.
